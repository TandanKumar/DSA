```python
s1 = [
    [4, 9, 2],
    [3, 5, 7],
    [8, 1, 5]
]

s2 =    [
    [5, 3, 4],
    [1, 5, 8],
    [6, 4, 2]
]

s3 = [
    [4, 8, 2],
    [4, 5, 7],
    [6, 1, 6]
]

lst_s = [s1,s2,s3]
```

```python
def formingMagicSquare(s):
    row_sum = [0,0,0]
    col_sum = [0,0,0]
    diag_sum = [0,0]
    n=2
    for i in range (3):
        r_sum =0
        for j in range(3):
            r_sum += s[i][j]
            col_sum[j]+=s[i][j]
            if i == j :
                #print(f"Inside the first if --- {s[i][j]}")
                diag_sum[0] += s[i][j]
            if n-i == j:
                diag_sum[1] += s[i][j]
        row_sum[i] = (r_sum)
    return(row_sum,col_sum,diag_sum)
            

```
