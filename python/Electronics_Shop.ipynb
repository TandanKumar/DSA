# Brute Force Approach

```python
def getMoneySpent(keyboards, drives, b):
    max_price = -1
    for ele in (keyboards):
        for drive in (drives):
            local_sum = ele + drive
            if local_sum > max_price and local_sum <= b :
                max_price = local_sum
    return max_price
```

# TWO Pointer Approach

```python
def getMoneySpent(keyboards, drives, b):
    keyboards.sort()
    drives.sort(reverse=True)
    max_price = -1
    i,j = 0,0
    while i < len(keyboards) and j < len(drives):
        total = keyboards[i] + drives[j]
        if total > b:
            j+=1
        else: 
            max_price = max(max_price , total )
            i+=1
    return max_price
```
