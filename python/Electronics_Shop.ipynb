# Brute Force Approach

```python
def getMoneySpent(keyboards, drives, b):
    max_price = -1
    for ele in (keyboards):
        for drive in (drives):
            local_sum = ele + drive
            if local_sum > max_price and local_sum <= b :
                max_price = local_sum
    return max_price
```
